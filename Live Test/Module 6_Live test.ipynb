```python
import pandas as pd
import numpy as np

data = {
    'A': [21, np.nan, 22, 4, np.nan],
    'B': [31, 32, 33, 34, 35],
    'C': [np.nan, np.nan, np.nan, np.nan, np.nan],
    'D': [46, 47, np.nan, 48, 49]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Finding total NaN values in the DataFrame
print("\nTotal NaN values in the DataFrame:", df.isna().sum().sum())

# Forward Fill
df_ffill = df.fillna(method='ffill')
print("\nDataFrame after Forward Fill:\n", df_ffill)

# Backward Fill
df_bfill = df.fillna(method='bfill')
print("\nDataFrame after Backward Fill:\n", df_bfill)

# Zero Fill
df_zero_fill = df.fillna(0).astype(int)
print("\nDataFrame after Zero Fill:\n", df_zero_fill)
```
