```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("shop data.csv")
```

```python
df.head()
```

```python
x = df.drop('buys', axis=1)
```

```python
x
```

```python
y = df.buys
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_x = LabelEncoder()
x = x.apply(le_x.fit_transform)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state=2)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dect = DecisionTreeClassifier(max_depth=5)
dect.fit(xtrain, ytrain)
```

```python
dect.predict(xtest)
```

```python
ytest
```

```python
dect.score(xtest, ytest)
```

```python
from sklearn.linear_model import LogisticRegression
logit_model = LogisticRegression()
logit_model.fit(xtrain, ytrain)
logit_model.predict(xtest)
logit_model.score(xtest, ytest)
```

```python
from sklearn.preprocessing import MinMaxScaler
MMS = MinMaxScaler()
x_scaled = MMS.fit_transform(xtrain)
x_scaled
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(xtrain, ytrain)
knn_model.predict(xtest)
knn_model.score(xtest, ytest)
```

```python
knn_model.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, knn_model.predict(xtest)))
```
