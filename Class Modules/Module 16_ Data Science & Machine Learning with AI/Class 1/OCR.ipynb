```python
import cv2
import pytesseract
import re

# Load image
image_path = 'C:/Ostad/OCR/WhatsApp Image 2025-06-26 at 22.49.51_7a6a130e.jpg'
image = cv2.imread(image_path)

# Convert image to RGB (pytesseract prefers RGB)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
# OCR processing
text = pytesseract.image_to_string(image_rgb)

# Print full OCR text (for debugging)
print("Full OCR Output:\n", text)
```

```python
type(text)
```

```python
import pandas as pd

lines = text.split('\n')
lines = [line.strip() for line in lines if line.strip() != '']  # Remove empty lines
df = pd.DataFrame(lines, columns=['Extracted Text'])
```

```python
df
```

```python
# Function to find target and extract next value
def extract_after_keyword(df, keyword):
    for i in range(len(df) - 1):
        if keyword.lower() in df.loc[i, 'Extracted Text'].lower():
            return df.loc[i + 1, 'Extracted Text'].strip()
    return None

# Extract values
name = extract_after_keyword(df, 'Name')
father_name = extract_after_keyword(df, "Father's Name")
mother_name = extract_after_keyword(df, "Mother's Name")

# Put results in new DataFrame
result_df = pd.DataFrame({
    'Field': ['Name', "Father's Name", "Mother's Name"],
    'Value': [name, father_name, mother_name]
})

print(result_df)
```
