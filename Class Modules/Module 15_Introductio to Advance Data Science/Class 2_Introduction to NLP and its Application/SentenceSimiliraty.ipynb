```python
import pandas as pd
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
w1 = 'pink'
w2 = 'red'

w1 = nlp.vocab[w1]
w2 = nlp.vocab[w2]
```

```python
w1.similarity(w2)
```

```python
data = pd.read_excel("Sentence.xlsx", 'Sheet1')
data
```

```python
ref_sent = data.iloc[1][1]
ref_sent
```

```python
ref_sent_vec = nlp(ref_sent)
ref_sent_vec
```

```python
type(ref_sent_vec)
```

```python
all_docs = [nlp(row) for row in data['Message']]
```

```python
all_docs
```

```python
type(all_docs[0])
```

```python
len(all_docs)
```

```python
sims = []
doc_id = []
for i in range(len(all_docs)):
    sim = all_docs[i].similarity(ref_sent_vec)
    sims.append(sim)
    doc_id.append(i)
```

```python
doc_id
```

```python
sims_docs = pd.DataFrame(list(zip(doc_id, sims)), columns = ['doc_id', 'sims'])
```

```python
list(zip("AB", "10"))
```

```python
sims_docs
```

```python
sims_docs_sorted = sims_docs.sort_values(by = 'sims', ascending=False)
sims_docs_sorted.head()
```

```python
sims_docs_sorted['doc_id'][0:5]
```

```python
top5_sim_doc = data.iloc[sims_docs_sorted['doc_id'][0:5]]
top5_sim_doc
```

```python
top_sims_score = pd.concat([top5_sim_doc, sims_docs_sorted['sims'][0:5]], axis=1)
top_sims_score
```
