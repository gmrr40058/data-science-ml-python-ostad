```python
maxdate2['InvoiceAge'] = (currentDate - maxdate2['InvoiceDate']).dt.days
maxdate2
```

```python
recency = maxdate2.drop('InvoiceDate', axis=1)
recency
```

```python
freq = df.drop_duplicates(subset='InvoiceNo')
freq
```

```python
freq = freq.groupby(['CustomerID']).count()[['InvoiceNo']]
freq
```

```python
df['Total'] = df['Quantity'] * df['UnitPrice']
df
```

```python
money = df.groupby(['CustomerID'])[['Total']].sum()
money
```

```python
df[df['CustomerID']==12346.0]
```

```python
RFM = pd.concat([recency, freq, money], axis=1)
RFM
```

```python
RFM.columns = ['Recency', 'Frequency', 'Monetary']
RFM
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(RFM)
scaled
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(scaled)
```

```python
kmeans.labels_
```

```python
RFM['Cluster'] = (kmeans.labels_)
RFM
```

```python
RFM['Cluster'].unique()
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111,projection='3d', label='bla')
ax.scatter(RFM['Recency'], RFM['Frequency'], RFM['Monetary'], s=40, c=RFM['Cluster'], marker='o', cmap='Accent')
ax.set_title('The plot of cluster')
plt.show()
```

```python
def mapping_func(row):
    if row['Cluster'] == 0:
        return 'Diamond'
    elif row['Cluster'] == 1:
        return 'Gold'
    else:
        return 'Silver'
    


```

```python
RFM['Group'] = RFM.apply(mapping_func, axis=1)
RFM
```
